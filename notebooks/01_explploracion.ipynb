# Exploración del dataset Heart Disease UCI

En este notebook se realiza el análisis exploratorio mínimo:
- Vista general de los datos
- Distribución de la variable objetivo
- Algunas relaciones simples entre variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")

df = pd.read_csv("../data/raw/heart.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["target"].value_counts(), df["target"].value_counts(normalize=True)
```

```python
df["target"].value_counts().plot(kind="bar")
plt.title("Distribución de la variable objetivo")
plt.xlabel("target (0 = sano, 1 = riesgo)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
sns.boxplot(x="target", y="age", data=df)
plt.title("Edad vs riesgo")
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Matriz de correlaciones")
plt.show()
```
